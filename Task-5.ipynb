```python
# ---------------------------
# Task 5: Exploratory Data Analysis (EDA) - Titanic Train Dataset
# ---------------------------

# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (replace 'train.csv' with your actual path if needed)
df = pd.read_csv("train.csv")

# ---------------------------
# Basic Info
# ---------------------------
print("Dataset Shape:", df.shape)
print("\n--- Dataset Info ---")
print(df.info())

print("\n--- First 5 Rows ---")
print(df.head())

# ---------------------------
# Summary Statistics
# ---------------------------
print("\n--- Summary Statistics ---")
print(df.describe(include='all'))

# ---------------------------
# Value Counts for each categorical column
# ---------------------------
print("\n--- Value Counts ---")
for col in df.select_dtypes(include=['object']):
    print(f"\nColumn: {col}")
    print(df[col].value_counts())

# ---------------------------
# Missing Values
# ---------------------------
print("\n--- Missing Values ---")
print(df.isnull().sum())

# ---------------------------
# Visualization - Pairplot (only numeric columns to avoid errors)
# ---------------------------
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.pairplot(numeric_df, diag_kind='kde')
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()

# ---------------------------
# Heatmap - Correlation
# ---------------------------
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# ---------------------------
# Histograms
# ---------------------------
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Numerical Features")
plt.show()

# ---------------------------
# Boxplots
# ---------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x='Sex', y='Age', data=df)
plt.title("Age Distribution by Sex")
plt.show()

# Survival rate by class
plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()

# ---------------------------
# Scatterplots
# ---------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Sex', data=df)
plt.title("Age vs Fare by Sex")
plt.show()

# ---------------------------
# Observations Example (you will add more after analysis)
# ---------------------------
print("\nObservations:")
print("1. Majority of passengers are male.")
print("2. Most passengers are in Pclass 3.")
print("3. Fare distribution is right-skewed with some high outliers.")
print("4. Age has missing values that need imputation.")
print("5. Correlation shows Fare and Pclass are inversely related.")

# ---------------------------
# Summary of Findings
# ---------------------------
print("\nSummary:")
print("Survival is influenced by passenger class, sex, and fare. Women and children had higher survival rates, and passengers in higher classes were more likely to survive.")
```
